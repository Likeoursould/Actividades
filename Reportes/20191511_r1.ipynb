1. Derive y explique la funci√≥n de demanda de Consumo. Replique el gr√°fico siguiente:

Encuentre el nivel de Ingreso de equilibrio 

Entendemos al modelo Ingreso-Gasto Keynesiano como un gr√°fico que expone un punto de encuentro entre la Demanda Agregada (DA) y el Ingreso agregrada (Y) que necesita la econom√≠a para estar en equilibrio, entendiendo a una f√≥rmula:

$$DA=Y$$

Antes de analizar el punto de equilibrio, es importante denominar variables:

Consumo: C = Co + bYd

Inversi√≥n: I = Io ‚àí hr

Gasto del Gobierno: G = Go

Tributaci√≥n: T = tY

Exportaciones: X = Xo

Importaciones: M = mYd

La Demanada Agregada significa la suma de todo lo anterior

Demanda Agregada: 

$$DA = C + I + G + X ‚àí M$$

Adem√°s sabemos que 
                  $$Y^d=(1+t)$$

Reemplazamos


$$DA=Co+bYd+Io-hr+Go+tY+Xo+mYd$$
$$Da=Co+Io+Go+Xo-hr+Y(b-m)(1-t)$$
     
Para encontrar la pendiente que nos dar√° el equilibrio debemos sintetizar la anterior formular, en donde:
$$ùê∑ùê¥ = ùõº0 + ùõº1ùëå$$

Nuestro intercepto estar√° determinada por : $$ùõº0= (C0 + I0 + G0 + X0 ‚àí hr)$$
Nuestra pendiente estar√° determinada por:   $$ùõº1= (b ‚àí m)(1 ‚àí t)$$
               
Ahora, considerando el equilibrio entre $$Y=DA$$, el punto de equilibrio estar√≠a conformado:

$$ Y^e = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ Y^e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Donde el multiplicador keynesiano es:

$$k= \frac{1}{1 - (b - m)(1 - t)}$$

Antes de analizar el punto de equilibrio, es importante denominar variables:

Consumo: C = Co + bYd

Inversi√≥n: I = Io ‚àí hr

Gasto del Gobierno: G = Go

Tributaci√≥n: T = tY

Exportaciones: X = Xo

Importaciones: M = mYd

La Demanada Agregada significa la suma de todo lo anterior

Demanda Agregada: 

$$DA = C + I + G + X ‚àí M$$

$$ DA = Œ±_0 + Œ±_1Y $$

Donde $ Œ±_0 = (C_0 + I_0 + G_0 + X_0 -hr)$ es el intercepto y $ Œ±_1 = [(b - m)(1 - t) - g] $ es la pendiente de la funci√≥n

Ahora, considerando la condici√≥n de equilibrio $Y = DA$, la ecuaci√≥n del ingreso de equilibrio a corto plazo es:

$$ Y^e = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$


Analice una Pol√≠tica Fiscal Expansiva con aumento del Gasto del Gobierno . El an√°lisis debe ser intuitivo, matem√°tico y gr√°fico.

$$ ‚ÜëGo ‚Üí ‚ÜëDA ‚Üí DA > Y ‚Üí ‚ÜëY $$

Matem√°tica:

```python

Co, Io, Go, Xo, h, r, b, m, t, g = symbols('Co Io Go Xo h r b m t g')

f = (Co + Io + Go + Xo - h*r)/(1 - (b - m)*(1 - t) + g)

df_Go = diff(f, Go) 
df_Go #‚àÜY/‚àÜGo
```

La pendiente es representada por el diferencial de $ ‚àÜGo/‚àÜY $: $ \frac{1}{g-(1-t)(b-m)+1} $ y es positivo

```python
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
g = 0.2

Y = np.arange(Y_size)


def DA_C(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_C = (Co  + Io + Go + Xo - h*r) + [(b-m)*(1-t)-g]*Y
    return DA_C

DA_Cont = DA_C(Co, Io, Go, Xo, h, r, b, m, t, g, Y)


Go = 100

def DA_C(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_C = (Co  + Io + Go + Xo - h*r) + [(b-m)*(1-t)-g]*Y
    return DA_C

DA_C_G = DA_C(Co, Io, Go, Xo, h, r, b, m, t, g, Y)
```

```python

def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
       raise Exception('lines do not intersect')

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

 
A = [DA_Cont[0], Y[0]] 
B = [DA_Cont[-1], Y[-1]]

C = [L_45[0], Y[0]] 
D = [L_45[-1], Y[-1]] 



intersec = line_intersection((A, B), (C, D))
intersec 
```
