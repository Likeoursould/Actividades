2. Derive la oferta real de dinero y explique cada uno de sus componentes.

La oferta real de dinero se puede explicar desde la siguiente fórmula, este se consigue entre la división de la oferta nominal del dinero ($M^s$), que es nuestra variable exógena y conforma parte de los instrumentos de política monetaria entre lo que conocemos como el nivel general de los precios ($P$) que funciona como constante y denominador, además, es importante aclarar que la función es dada en corto plazo, pues se supone que el nivel de precios ya está dado. 



$$ \frac{M^s}{P} = \frac{M^s_o}{P} $$

3. Derive la demanda real de dinero. Explique qué papel cumplen los parametros "k" y "j"

La demanda de real de dinero funciona porque existe los consumidores como las familias y las empresas que demandas este dinero, ha esta necesidad de dinero podemos encontrar las siguiente funciones.
Por un lado las demandas de dinero por motivos de transacción y precaución, en donde por estos motivos las familias o empresas se endeudan; sin embargo, estas deben tener la capacidad de pago, esto dependiendo de sus ingresos (Y)

$$ L_1 = kY $$

Por otra parte, existe la demanda de dinero por especulación, en donde los activos financieramente prefieren tener su dinero a modo de dinero y no de bono dependiendo de la tasa de interés del momento 

$$ L_2 = -ji $$

Teniendo estos dos conceptos en mente, podemos determinar que nuestra función de demanda real de dinero es

$$ L = L_1 + L_2 $$

$$ M^d = kY -ji $$

4. Asumiendo que no hay inflación podemos asumir que . Escriba en terminos reales la eucacion de equilibrio en el mercado de dinero

Para encontrar el eequilibrio del mercado de dinero, asumiendo que no hay inflación, este se deriva del equilibrio de la Oferta de dinero y la Demanda de dinero 
$$ M^s = M^d $$

Además nuestra formula de Demanda de Dinero está especificada por la diferencia entre demanda por preocupaciones y transacción menos la demanda por especulación 
$$ M^d = kY -ji $$

Entonces nuestra fórmula de equilibrio sería de la siguiente manera
$$ \frac{M^s}{P} = kY - ji $$



Suponiendo que la inflación es cero, no existe un gran diferencia enter la tasa de interés nominal ($i$) y la tasa de interés real ($r$), entonces la ecuacíon sería de la siguiente manera:

$$ \frac{M^s}{P} = kY - jr $$


5. Grafique el equilibrio en el mercado de dinero.

```python

r_size = 100

k = 0.5
j = 0.2                
P  = 10 
Y = 35

r = np.arange(r_size)


def Ms_MD(k, j, P, r, Y):
    Ms_MD = P*(k*Y - j*r)
    return Ms_MD

Ms_MD = Ms_MD(k, j, P, r, Y)
```

```python
    
fig, ax1 = plt.subplots(figsize=(10, 8))

ax1.set(title="Equilibrio en el Mercado de Dinero", xlabel=r'M^s / P', ylabel=r'r')
ax1.plot(Ms_MD, label= '$L_0$', color = '#E539C3')

ax1.axvline(x = 45,  ymin= 0, ymax= 1, color = "grey")
ax1.axhline(y=85, xmin= 0, xmax= 0.46, linestyle = ":", color = "black")
ax1.text(0, 89, "$r_0$", fontsize = 12, color = 'black')

ax1.yaxis.set_major_locator(plt.NullLocator())   
ax1.xaxis.set_major_locator(plt.NullLocator())

ax1.legend()

plt.show()
```

$$ \frac{M^s}{P} = kY - jr $$

Moviendo en función de $r$, la curva de LM 
$$ r = -\frac{1}{j}\frac{M_o^s}{P_o} + \frac{k}{j}Y $$

2. ¿Cuál es el efecto de una disminución en la Masa Monetaria ∆Ms<0  ? Explica usando la intuición y gráficos

Intuicición:

Teninedo en mente la formula de equilibrio, se derivaría de la siguiente manera:
$$ Ms↓ → Ms < Md → r↑ $$

Graficamente:

```python

Y_size = 100

k = 2
j = 1                
Ms = 900             
P  = 20               

Y = np.arange(Y_size)

# Ecuación

def i_LM( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i = i_LM( k, j, Ms, P, Y)

#--------------------------------------------------

Ms = 200

def i_LM_Ms( k, j, Ms, P, Y):
    i_LM = (-Ms/P)/j + k/j*Y
    return i_LM

i_Ms = i_LM_Ms( k, j, Ms, P, Y)
```

```python

y_max = np.max(i)
v = [0, Y_size, 0, y_max]   
fig, ax = plt.subplots(figsize=(10, 8))


ax.plot(Y, i, label="LM", color = 'black')
ax.plot(Y, i_Ms, label="LM_Ms", color = '#C5DE0D', linestyle = 'dashed')


plt.text(48, 74, '∆$M^s$', fontsize=12, color='black')
plt.text(49, 68, '←', fontsize=15, color='grey')


ax.set(title = "Contracción en la Masa Monetaria $(M^s)$", xlabel=r'Y', ylabel=r'r')
ax.legend()

plt.show()
```
